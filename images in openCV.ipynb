```python
img = cv2.imread("DATA/00-puppy.jpg")
plt.imshow(img)
```

```python
new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(new_img)
```

```python
img_gray = cv2.imread("DATA/00-puppy.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
resize_img = cv2.resize(img,(800,100))
```

```python
plt.imshow(resize_img)
```

```python
v_ratio=0.5
h_ratio=0.8
```

```python
resize_img = cv2.resize(img,(0,0),img,v_ratio,h_ratio)
```

```python
plt.imshow(resize_img)
```

```python
flip_img = cv2.flip(new_img,0)
```

```python
plt.imshow(flip_img)
```

```python
flip_img = cv2.flip(new_img,1)
plt.imshow(flip_img)
```

```python
flip_img = cv2.flip(new_img,-1)
plt.imshow(flip_img)
```
